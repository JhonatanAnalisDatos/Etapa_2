```python
#Cargamos a nuestro entorno de trabajo las librerias necesarias.

import matplotlib.pyplot as plt
from sklearn import linear_model # usamos sklear para saber los valores optimos.
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
#Cargamos la informacion del CSV en una variable llamda datos.

datos = pd.read_csv("C:/INGENIERIA DE SISTEMAS/SEXTO SEMESTRE/Analisis de Datos/1 UNIDAD/Tarea 2/DATOS/data.csv", sep=",")
```

```python
#Visualisamos los datos del archivo CSV data

datos
```

```python
#Realizamos la grafica de dispersion y visualisamos

datos.plot.scatter(x="metro", y="precio")
plt.show()
```

```python
#Cremos una instancia del objeto LinearRegression de la biblioteca Sklearn

regresion = linear_model.LinearRegression()
```

```python
#Dividimos las variables Independientes (metros) y Dependiente (precio)

x = datos["metro"].values.reshape(-1, 1)
y = datos["precio"].values
```

```python
#Ajustamos el modelo a los datos

modelo = regresion.fit(x,y)
```

```python
#Imprimimos la Interseccion

print("Interseccion (b)", modelo.intercept_)

#Imprimimos la Pendiente

print("Pendiente (m)", modelo.coef_)
```

```python
#Creamos el modelo regresion lineal simple

plt.title('MODELO REGRESION LINEAL SIMPLE')
plt.scatter(x, y, color= "blue", label="Datos")
plt.plot(x, modelo.predict(x), color="red", label="Línea de regresión")
plt.xlabel("METROS", color = "red")
plt.ylabel("PRECIOS", color = "red")
plt.legend()
plt.show()
```
